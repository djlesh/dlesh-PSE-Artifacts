## Question 1
Using the Pyplot library from Matplotlib graph the average temperature for the following months on a line graph. This should be similar to the examples we looked at with 2 lists of numbers. Ensure that your months are on the x-axis. Add an appropriate title and X/Y labels to your graph.


```python
months = ["June", "July", "August", "September"]
average_temperature = [84, 91, 88, 78]
```

```python
import matplotlib.pyplot as plt

months = ["June", "July", "August", "September"]
average_temperature = [84, 91, 88, 78]

fig, ax = plt.subplots()
ax.plot(months, average_temperature, linewidth=3)


# Set chart title and label axes.
ax.set_title("Average Temperatures", fontsize = 24)
ax.set_xlabel("Month", fontsize = 14)
ax.set_ylabel("Temperature", fontsize = 14)

# Set size of tick labels.
ax.tick_params(axis='both', labelsize=14)

plt.show()
```
