```python
# Import Packages
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import dataset via df function
MLBHitters_df = pd.read_csv("/Users/daniellesh/Desktop/OPIM 3223/Group Project Materials/MLBHitters.csv")
```

```python
# Displaying first five rows and last five rows of MLBHitters_df
display(MLBHitters_df.head())
display(MLBHitters_df.tail())

# Filtering for players with at least (>=) 30 HR in 2021 Season
TopHRHitters_df = MLBHitters_df[MLBHitters_df.HR >= 30]

# Redisplay first five rows and last five rows of TopHRHitters_df
display(TopHRHitters_df.head())
display(TopHRHitters_df.tail())
```

```python
# HR simple statistics (Home Runs)
display(TopHRHitters_df["HR"].describe())
```

```python
# OBP simple statistics (On-base percentage)
display(TopHRHitters_df["OBP"].describe())
```

```python
# SLG simple statistics (Slugging -> # of bases per at-bat)
display(TopHRHitters_df["SLG"].describe())
```

```python
# OPS simple statistics (On-base + slugging)
# OPS = OBP + SLG
display(TopHRHitters_df["OPS"].describe())
```

```python
# AVG simple statistics (Batting Average)
display(TopHRHitters_df["AVG"].describe())
```

```python
# Prevents conflict between pandas and matplotlib
pd.plotting.register_matplotlib_converters()

# CSV derived graph
filename = 'MLBHitters.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    players, OPS, OBP = [], [], []
    for row in reader:
        if int(row[10]) >= 30:
            player = str(row[0])
            OPS1 = float(row[22])
            OBP1 = float(row[20])
            players.append(player)
            OPS.append(OPS1)
            OBP.append(OBP1)

# Plot OPS and On-Base %
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(players, OPS , c='red', alpha=0.5, label = 'OPS')
ax.plot(players, OBP, c='blue', alpha=0.5, label = 'OBP')

# Fill Plot
ax.fill_between(players, OPS, OBP, facecolor='green', alpha=0.5)

# Format plot
title = "Comparing On-Base % + Slugging (OPS) & On-Base % (OBP)"
ax.set_title(title, fontsize=18)
ax.set_xlabel('Players', fontsize=16)
ax.set_ylabel("Percentage (%)", fontsize=16)
ax.tick_params(axis='x', which='major', rotation=90, labelsize=12.5)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Prevents conflict between pandas and matplotlib
pd.plotting.register_matplotlib_converters()

# Set Player, On-Base %, and Batting Average
Player = TopHRHitters_df['Player']
On_Base_PCT = TopHRHitters_df['OBP']
Batting_Average = TopHRHitters_df['AVG']

# Plot On-Base % and Batting Average
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.plot(Player, On_Base_PCT , c='red', alpha=0.5, label = 'OBP')
ax.plot(Player, Batting_Average, c='green', alpha=0.5, label = 'AVG')

# Fill Plot
ax.fill_between(Player, On_Base_PCT, Batting_Average, facecolor='navy', alpha=0.5)

# Format plot
title = "Comparing Batting Average (AVG) & On-Base % (OBP)"
ax.set_title(title, fontsize=18)
ax.set_xlabel('Players', fontsize=16)
ax.set_ylabel("Percentage (%)", fontsize=16)
ax.tick_params(axis='x', which='major', rotation=90, labelsize=12.5)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Top HR Hitters by Position (Histogram)
x = TopHRHitters_df["Pos"]
plt.style.use('seaborn')
fig, ax = plt.subplots()

ax.hist(x, bins=7, linewidth=3, edgecolor="white")

ax.set(xlim=(-1, 7), xticks=np.arange(-1, 7),
       ylim=(0, 13), yticks=np.linspace(0, 13, 14))

# Format plot
title = "Number of Home Run Hitters by Position"
ax.set_title( title, fontsize=18)
ax.set_xlabel('Position', fontsize=16)
ax.set_ylabel("Number of Players", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```

```python
# Comparing Homeruns and Strikeouts (Scatterplot)
x = TopHRHitters_df["HR"]
y = TopHRHitters_df["SO"]

sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(28, 50), xticks=np.arange(28, 50, 2),
       ylim=(85, 200), yticks=np.arange(85, 200, 10))

# Format plot
title = "Comparing Home Runs & Strikeouts"
ax.set_title( title, fontsize=18)
ax.set_xlabel('Number of Home Runs', fontsize=16)
ax.set_ylabel("Number of Strikeouts", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```

```python
# Number of Home Runs by Player
df_sorted = TopHRHitters_df.sort_values('HR', ascending = False)

# Set Player, Home Runs
Player = df_sorted['Player']
Homers = df_sorted['HR']

fig, ax = plt.subplots()

ax.bar(Player, Homers, width=1, edgecolor="white", linewidth=0.7)

# Format Plot
title = 'Number of Home Runs by Player'
ax.set_title(title, fontsize=18)
ax.set(xlim=(-1, 38), xticks=np.arange(-1, 38),
       ylim=(0, 50), yticks=np.arange(0, 50, 5))
ax.set_xlabel("Players", fontsize=16)
ax.set_ylabel("Number of Home Runs", fontsize=16)
ax.tick_params(axis='x', which='major', rotation=90, labelsize=12.5)
plt.show()
```
