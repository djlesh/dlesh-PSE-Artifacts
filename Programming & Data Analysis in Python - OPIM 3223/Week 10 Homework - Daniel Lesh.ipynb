```python
import csv
from datetime import datetime

from matplotlib import pyplot as plt

import csv

filename = '/Users/daniellesh/Desktop/OPIM 3223/Data/2010_Northridge_CA_Weather.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
      
    # Get dates and high temperatures from this file.
    dates, avg_temps = [], []
    for row in reader:
        # We select index 2 which contains the date information
        # We use the datatime module to convert this data into a datetime object
        current_date = datetime.strptime(row[5], '%m/%d/%y')
        dates.append(current_date)
        avg_temp = float(row[8])
        avg_temps.append(avg_temp)

# Plot the average temperatures.
plt.style.use('seaborn')
fig, ax = plt.subplots()
# We plot our date values in addition to the high temperatures
ax.plot(dates, avg_temps, c='navy')

# Format plot.
ax.set_title("Daily Average Temperatures, January 2010 - December 2010\n", fontsize = 18)
ax.set_xlabel('\n Month and Year', fontsize=16)
# fig.autofmt_xdate() automatically formats our dates to prevent overlap and shows them diagonally
fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)\n", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```
